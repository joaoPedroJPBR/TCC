```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./diabetes_prediction_dataset.csv", sep=",")
```

```python
df
```

```python
# Criando um grÃ¡fico de dispersÃ£o para idade e nÃ­vel de glicose no sangue
plt.scatter(df['age'], df['blood_glucose_level'], c=df['diabetes'], cmap='viridis')
plt.colorbar(label='Diabetes')
plt.xlabel('Idade')
plt.ylabel('NÃ­vel de Glicose no Sangue')
plt.title('DispersÃ£o entre Idade e NÃ­vel de Glicose no Sangue')
plt.show()
```

## Histograma da Idade

```python
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Idade')
plt.ylabel('FrequÃªncia')
plt.title('DistribuiÃ§Ã£o da Idade na Amostra')
plt.show()
```

## GrÃ¡fico de Barras para HipertensÃ£o e DoenÃ§a CardÃ­aca

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

df['hypertension'].value_counts().plot(kind='bar', ax=axes[0], color='lightcoral')
axes[0].set_title('IncidÃªncia de HipertensÃ£o')
axes[0].set_xticklabels(['Sem HipertensÃ£o', 'Com HipertensÃ£o'], rotation=0)

df['heart_disease'].value_counts().plot(kind='bar', ax=axes[1], color='lightgreen')
axes[1].set_title('IncidÃªncia de DoenÃ§a CardÃ­aca')
axes[1].set_xticklabels(['Sem DoenÃ§a CardÃ­aca', 'Com DoenÃ§a CardÃ­aca'], rotation=0)

plt.tight_layout()
plt.show()
```

## GrÃ¡fico de Pizza para HistÃ³rico de Fumo

```python
smoking_counts = df['smoking_history'].value_counts()
plt.pie(smoking_counts, labels=smoking_counts.index, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightgreen'])
plt.title('DistribuiÃ§Ã£o do HistÃ³rico de Fumo')
plt.show()
```

## Diagrama de DispersÃ£o para BMI e NÃ­vel de Glicose no Sangue

```python
plt.scatter(df['bmi'], df['blood_glucose_level'], c=df['diabetes'], cmap='viridis')
plt.colorbar(label='Diabetes')
plt.xlabel('Ãndice de Massa Corporal (BMI)')
plt.ylabel('NÃ­vel de Glicose no Sangue')
plt.title('DispersÃ£o entre BMI e NÃ­vel de Glicose no Sangue')
plt.show()
```

```python
# Conta o nÃºmero de pessoas com diabetes
contagem_diabetes = df['diabetes'].sum()

# Conta o nÃºmero de pessoas sem diabetes
contagem_sem_diabetes = len(df) - contagem_diabetes

# Cria um grÃ¡fico de barras
rÃ³tulos = ['Sem Diabetes', 'Com Diabetes']
contagens = [contagem_sem_diabetes, contagem_diabetes]

plt.bar(rÃ³tulos, contagens, color=['blue', 'orange'])
plt.xlabel('Diabetes')
plt.ylabel('NÃºmero de Pessoas')
plt.title('NÃºmero de Pessoas com e sem Diabetes')
plt.show()
```
