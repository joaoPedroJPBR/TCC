```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./diabetes_prediction_dataset.csv", sep=",")
```

```python
df
```

```python
# Criando um gráfico de dispersão para idade e nível de glicose no sangue
plt.scatter(df['age'], df['blood_glucose_level'], c=df['diabetes'], cmap='viridis')
plt.colorbar(label='Diabetes')
plt.xlabel('Idade')
plt.ylabel('Nível de Glicose no Sangue')
plt.title('Dispersão entre Idade e Nível de Glicose no Sangue')
plt.show()
```

## Histograma da Idade

```python
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Distribuição da Idade na Amostra')
plt.show()
```

## Gráfico de Barras para Hipertensão e Doença Cardíaca

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

df['hypertension'].value_counts().plot(kind='bar', ax=axes[0], color='lightcoral')
axes[0].set_title('Incidência de Hipertensão')
axes[0].set_xticklabels(['Sem Hipertensão', 'Com Hipertensão'], rotation=0)

df['heart_disease'].value_counts().plot(kind='bar', ax=axes[1], color='lightgreen')
axes[1].set_title('Incidência de Doença Cardíaca')
axes[1].set_xticklabels(['Sem Doença Cardíaca', 'Com Doença Cardíaca'], rotation=0)

plt.tight_layout()
plt.show()
```

## Gráfico de Pizza para Histórico de Fumo

```python
smoking_counts = df['smoking_history'].value_counts()
plt.pie(smoking_counts, labels=smoking_counts.index, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightgreen'])
plt.title('Distribuição do Histórico de Fumo')
plt.show()
```

## Diagrama de Dispersão para BMI e Nível de Glicose no Sangue

```python
plt.scatter(df['bmi'], df['blood_glucose_level'], c=df['diabetes'], cmap='viridis')
plt.colorbar(label='Diabetes')
plt.xlabel('Índice de Massa Corporal (BMI)')
plt.ylabel('Nível de Glicose no Sangue')
plt.title('Dispersão entre BMI e Nível de Glicose no Sangue')
plt.show()
```

```python
# Conta o número de pessoas com diabetes
contagem_diabetes = df['diabetes'].sum()

# Conta o número de pessoas sem diabetes
contagem_sem_diabetes = len(df) - contagem_diabetes

# Cria um gráfico de barras
rótulos = ['Sem Diabetes', 'Com Diabetes']
contagens = [contagem_sem_diabetes, contagem_diabetes]

plt.bar(rótulos, contagens, color=['blue', 'orange'])
plt.xlabel('Diabetes')
plt.ylabel('Número de Pessoas')
plt.title('Número de Pessoas com e sem Diabetes')
plt.show()
```
