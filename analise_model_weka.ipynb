```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/diabetes_prediction_weka.csv", sep=",")
```

```python
df
```

```python
# Conta o número de pessoas com diabetes
contagem_diabetes = df['class'].sum()

# Conta o número de pessoas sem diabetes
contagem_sem_diabetes = len(df) - contagem_diabetes

# Cria um gráfico de barras
rótulos = ['Sem Diabetes', 'Com Diabetes']
contagens = [contagem_sem_diabetes, contagem_diabetes]

plt.bar(rótulos, contagens, color=['green', 'orange'])
plt.xlabel('Diabetes')
plt.ylabel('Número de Pessoas')
plt.title('Número de Pessoas com e sem Diabetes após balanceamento da base de dados')
plt.show()
```
