```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/diabetes_prediction_weka.csv", sep=",")
```

```python
df
```

```python
# Conta o nÃºmero de pessoas com diabetes
contagem_diabetes = df['class'].sum()

# Conta o nÃºmero de pessoas sem diabetes
contagem_sem_diabetes = len(df) - contagem_diabetes

# Cria um grÃ¡fico de barras
rÃ³tulos = ['Sem Diabetes', 'Com Diabetes']
contagens = [contagem_sem_diabetes, contagem_diabetes]

plt.bar(rÃ³tulos, contagens, color=['green', 'orange'])
plt.xlabel('Diabetes')
plt.ylabel('NÃºmero de Pessoas')
plt.title('NÃºmero de Pessoas com e sem Diabetes apÃ³s balanceamento da base de dados')
plt.show()
```
