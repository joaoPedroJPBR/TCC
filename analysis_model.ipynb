```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/model_metrics.csv", sep=";")
```

```python
df
```

```python
# Criar o gráfico de barras
plt.bar(df["Modell"], df["Accuracy"], color=['blue', 'orange', 'green'])
plt.ylim(0.9, 1.0)  # Definir o intervalo do eixo y para melhor visualização
plt.xlabel('Modelo')
plt.ylabel('Acurácia')
plt.title('Acurácia dos Modelos')
plt.show()
```
